```python
query = "Como o Alarik usa S3"

# simula a KB
sections = {
    "Introdução": """
    Alarik usa S3 para armazenar arquivos.
    O S3 é utilizado para backup.
    """,

    "Arquitetura": """
    Alarik possui integração com Databricks.
    Alarik também usa Spark para processamento.
    """,

    "Banco de Dados": """
    PostgreSQL é usado para armazenar metadados.
    """
}

# 1) termos da query
terms = [t.lower() for t in query.split() if len(t) > 2]
print("TERMS:", terms)

for t in query.split():
    if len(t) > 2:
        print(t)
```

```python
hits = []
print("HITS INICIAL:", hits)

for title, body in sections.items():
    score = 0                          # zera pra CADA seção
    for t in terms:
        n = body.lower().count(t)      # quantas vezes 't' aparece nesta seção
        score = score + n              # vai acumulando
        print(f"  termo '{t}' aparece {n}x  -> score parcial {score}")
    print(f"SEÇÃO '{title}': score final = {score}\n")
    if score > 0:
        hits.append((score, body))
```

```python

# 2) hits
hits = []
print("HITS INICIAL:", hits)

# 3) scoring
for title, body in sections.items():
    score = sum(body.lower().count(t) for t in terms)
    print(score)

    if score:
        hits.append((score, body))

# 4) ordenação
hits.sort(reverse=True, key=lambda x: x[0])
print("\nHITS ORDENADOS:", hits)

# 5) retorno final
result = "\n\n---\n\n".join(b for _, b in hits[:3]) or "Nada encontrado. Use list_sections."

print("\nRESULTADO FINAL:\n")
print(result)
```
